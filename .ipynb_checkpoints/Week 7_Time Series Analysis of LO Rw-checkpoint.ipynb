```python
df['Loadout Date']=pd.to_datetime(df['Loadout Date'],format='%Y-%m-%d %H:%M:%S')
```

```python
df=df.sort_values(by='Loadout Date').set_index(keys='Loadout Date')
```

```python
df.head()
```

```python
df_row=df[['LO Row']]
```

### Analysis

```python
def plot_time_data(df,duration):
    func_df=df['LO Row'].resample(duration).sum()
    plt.figure(figsize=(18,6))
    plt.plot(func_df)
    plt.ylabel('LO Row')
    plt.title('Total LO Row vs Time ( '+duration+' )')
```

```python
plot_time_data(df_row,'D')
```

```python
plot_time_data(df_row,'5D')
```

```python
plot_time_data(df_row,'8H')
```
